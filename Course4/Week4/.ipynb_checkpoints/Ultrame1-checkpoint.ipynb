```python
import tensorflow as tf
import numpy as np
import matplotlib.image
```

```python
from tensorflow.python.platform import gfile
```

```python
graph_def = tf.GraphDef()
```

```python
with gfile.FastGFile('./models/20180402-114759.pb','rb') as f:
    graph_def.ParseFromString(f.read())
```

```python
input_var = tf.Variable(initial_value=np.zeros((1, 160, 160, 3)), dtype = 'float32')
```

```python
tf.import_graph_def(graph_def, input_map={"input:0": input_var}, name='mg')
```

```python
# images_placeholder = tf.get_default_graph().get_tensor_by_name("mg/input:0")
embeddings = tf.get_default_graph().get_tensor_by_name("mg/embeddings:0")
phase_train_placeholder = tf.get_default_graph().get_tensor_by_name("mg/phase_train:0")
```

```python
from matplotlib.pyplot import imshow, imsave, imread

image = matplotlib.image.imread('./David_R.png')
David_R = image.reshape(1, 160, 160, 3)
```

```python
imshow(image)
```

```python
def compute_embedding_cost(a_C, a_G):
    J_content = tf.reduce_mean(tf.square(a_C - a_G))/2
    return J_content
```

```python
david_B = matplotlib.image.imread('./David_B.png')
David_B = david_B.reshape(1, 160, 160, 3)
```

```python
imshow(david_B)
```

```python
def total_cost(Original, Target, alpha = 10, beta = 40):
    J = alpha*Original + beta*Target
    return J
```

```python
def generate_noise_image(content_image, noise_ratio = 0.6):
    """
    Generates a noisy image by adding random noise to the content_image
    """
    
    # Generate a random noise_image
    noise_image = np.random.uniform(-20, 20, (1, 160, 160, 3)).astype('float32')
    
    # Set the input_image to be a weighted average of the content_image and a noise_image
    input_image = noise_image * noise_ratio + content_image * (1 - noise_ratio)
    
    return input_image
```

```python
generated_image = generate_noise_image(david_B)
gen = generated_image.reshape(1,160,160,3)
imshow(generated_image[0])
```

```python
sess = tf.InteractiveSession()
```

```python
init_global = tf.global_variables_initializer()

sess.run(init_global)
sess.run(input_var.assign(David_R))
emb_T = sess.run(embeddings, feed_dict={phase_train_placeholder:False })
sess.run(input_var.assign(David_B))
emb_O = sess.run(embeddings, feed_dict={phase_train_placeholder:False })    
```

```python
emb_G = embeddings
J_t = compute_embedding_cost(emb_T, emb_G)
J_o = compute_embedding_cost(emb_O, emb_G)
J = total_cost(J_o, J_t)

optimizer = tf.train.AdamOptimizer(0.002)
train_step = optimizer.minimize(J)
```

```python
def save_image(path, image):    
    # Clip and Save the image
#     image = np.clip(image[0], 0, 255).astype('uint8')
    imsave(path, image)
```

```python
def model_nn(sess, input_image, num_iterations = 2000):
    
    sess.run(tf.global_variables_initializer())
    sess.run(input_var.assign(input_image))
    
    for i in range(num_iterations):
    
        sess.run(train_step, feed_dict={phase_train_placeholder:False})
        generated_image = sess.run(input_var, feed_dict={phase_train_placeholder:False})
        if i%200 == 0:
            Jt, Jc, Js = sess.run([J, J_t, J_o], feed_dict={phase_train_placeholder:False})
            print("Iteration " + str(i) + " :")
            print("total cost = " + str(Jt))
            print("original cost = " + str(Jc))
            print("target cost = " + str(Js))
            
            save_image("./output/" + str(i) + ".png", generated_image[0])
    
    # save last generated image
    save_image('./output/generated_image.jpg', generated_image[0])
    
    return generated_image
```

```python
model_nn(sess, David_B)
```
